### Wavelet Transformation:
Wavelet transforms are mathematical transformations used for analyzing and processing signals and time-series data. Unlike Fourier transforms, which decompose signals into sinusoids of varying frequencies, wavelet transforms decompose signals into wavelets, which are localized in both time and frequency domains. This localization property allows wavelet transforms to capture both high-frequency and low-frequency components of signals with better resolution compared to Fourier transforms.

```python
import numpy as np
import matplotlib.pyplot as plt
import pywt

# Generate time series data
np.random.seed(42)
t = np.arange(0, 10, 0.1)  # time points
n = len(t)
freq1 = 1  # frequency of the first sine wave
freq2 = 2  # frequency of the second sine wave
amplitude1 = 3  # amplitude of the first sine wave
amplitude2 = 2  # amplitude of the second sine wave
noise_amplitude = 1  # amplitude of the noise

# Generate two sine waves with different frequencies and add noise
time_series = amplitude1 * np.sin(2 * np.pi * freq1 * t) + amplitude2 * np.sin(2 * np.pi * freq2 * t) + noise_amplitude * np.random.randn(n)

# Perform Continuous Wavelet Transform
wavelet = 'morl'  # Morlet wavelet
scales = np.arange(1, 128)  # Scales for analysis
coefficients, frequencies = pywt.cwt(time_series, scales, wavelet)

# Plot the Continuous Wavelet Transform
plt.figure(figsize=(12, 6))
plt.imshow(np.abs(coefficients), extent=[0, len(time_series), scales[-1], scales[0]], aspect='auto', cmap='jet')
plt.colorbar(label='Magnitude')
plt.title('Continuous Wavelet Transform')
plt.xlabel('Time')
plt.ylabel('Scale')
plt.show()
```
